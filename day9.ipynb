```python
# 2. Create a class Shape with a method area(). Inherit classes Rectangle and Circle from it and override area().
import math
from abc import ABC,abstractmethod
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
```

```python
class Rectangle(Shape):
    def area(self,le,bre):
        self.ar=le*bre
        return self.ar
class Circle(Shape):
    def area(self,r):
        return math.PI*r**2
r=Rectangle()
print(r.area(3,5))
```

```python
# 3. Implement operator overloading using polymorphism: Create a class Book and override the + operator to add prices of two books.
class Book:
    def __init__(self,name,price):
        self.name=name
        self.price=price
    def __add__(self,other):
        return self.price + self.price
b=Book('vasu',344)
b2=Book('ls',342)
print(b+b2)
```

```python
# 4. Function polymorphism: Write a function multiply() that behaves differently when given integers, floats, or lists.

def multiply(a,b):
    return a*b
print(multiply(5,9))
print(multiply(5.86,9.82))
print(multiply([3,4,2,5,5],2))# for mutlplication here doing repetation
# Same function name, different behavior based on input type
```
