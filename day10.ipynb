```python
# logging decorator example
def log(func):
    def inner(*args,**kwargs):
        
            print(f"Calling function {func.__name__} with args:{args} and kwargs:{kwargs}")
            res=func(*args,**kwargs)
            print(f"function {func.__name__} returned {res}")
            return res
    return inner
@log
def add(a,b):
    return a+b
add(5,6)
    
    
```

```python
# 4.Measuering execution time 
import time
def benchmark(func):
    def inner(*args,**kwargs):
        
            st_time=time.time()
            res=func(*args,**kwargs)
            end_time=time.time()
            print(f"Execution time  {func.__name__} :{st_time-end_time} seconds")
            return res
    return inner
@log
@benchmark
def slow_fun(n):
    t=0
    for i in range(n):
        t=t+i
    return t
slow_fun(50)


```

```python
# how to pass the arguments for the decorators

# i want to print he particular hi as many times i want 
def repeater(n):
    def decorator(func):
        def inner(*args,**kwargs):
            for _ in range(n):
                func(*args,*kwargs)
        return inner
    return decorator
@repeater(3)
def say():
    print('hi')
say()
```
