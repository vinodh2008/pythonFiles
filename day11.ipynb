```python
# you have given an 8*8 chessboard 
# where cloumns are labeled from 'a'  to  'h'(left tp right)
# whereas rows are numbered from 1 to 8(bottom to up)
# your are given a starting position in chess notation (eg:- "b4")
# and list movements commands in the format 
# "u2" :-up two steps
# "d3" :-down three steps
# "r4" :- right four steps
# "l1" :- move left one step
#  your task is to navigate through the moves and returns  the final chess notation
#  if a move goes out of the bouuds,treturn the invalid
```

```python
def  chessboard_game(start,moves):
    col=start[0]
    row=int(start[1])# "4"
    
    for i in moves:
        dir=i[0]
        steps=int(i[1])
        if dir=="u":
            row=row+steps
        elif dir=="d":
            row=row-steps
        elif dir=="l":
            col=chr(ord(col)-steps)
        else:
            col=chr(ord(col)+steps)
        if (col>"h" or  col<"a") or (row<1 or row>8):
            return "invalid route"
    return f"{col}{row}"

    
    
        
            
   
chessboard_game("b4",["u2","d3","r4","l2"])
```

```python
# # reverse array by k times
#  brute force approach
k=2
p=[4,6,3,29,5]
for i in range(k):
    temp=p[len(p)-1]
    for i in range(len(p)-2,-1,-1):# for shifting element from 0 to n-1
        p[i+1]=p[i]
    p[i]=temp
    
print(p)
```

```python
#  this is optimal approach using the reverse overall list once ,then reverse 0 to kth elemements then reverse k+1 th element to  len(p)-1
def reverse(l):
    k=0
    m=len(l)-1
    for i in range(len(l)//2):# two pointer approach is considerable O(n)
        l[k],l[m]=l[m],l[k]
        k=k+1
        m=m-1
    # print(l)
    return l
# reverse([4,3,2,36,2])
l=reverse([4,6,3,26,3])
print(reverse(l[:2]))
print(reverse(l[2:]))

```
