 # diffence between the shallow copy and deepcopy
 shallow copy means duplicating the origainl list with new list but inner lists in the orginal list can't be duplicated 
 deep copy means duplicating the orginal list entirly with new list  ,also can be inner list also be dupliacted unlike shallow copy any modification to innerlist cannot be affected to the origainl list


```python
# example explanation
# 1. shallow copy:- creates new obj but not for innere elements in that original list
import copy# first import the copy module
org_list=[3,5,[5,3,2]]
new_list=copy.copy(org_list)
print(org_list,new_list)
org_list[2][1]='x'# here we modified the orgainl list after creating the copy for above org _list
# new_list=copy.copy(org_list)
print(org_list,new_list) #but still its relflecting the mdofication of the inner list can be reflected the duplicated list
# to overcome this problem the shallow copy is come into the pictire



```

```python
# example explanation
# 2.Deep copy :- create the new list like  shallow copy   but entire copy of org_list would be duplicated into the at new list and also inner list modification 
# can not be reflected to the new list
import copy# first import the copy module
org_list=[3,5,[5,3,2]]
new_list=copy.deepcopy(org_list)
print(org_list,new_list)
org_list[2][1]='x'# here we modified the orgainl list after creating the copy for above org _list
print(org_list,new_list)# even we modify the inner list can't be afftected to new list
```
