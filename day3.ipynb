```python
def upper_case(fucn):
    def wrapper():
        result=fucn()
        return result.upper()
    return wrapper
@upper_case
def my_string():
    return "Hello World"
print(my_string())
```

```python
# summation of all digits in a number
# take the number from the user
user_input=int(input("enter the number that you want to perform the all digits in given number"))
sum=0

n=user_input
# take the last digit from the one one by one
while(n>0): 
    rem=n%10
    sum=sum+rem
    n=n//10
print("sum of all digits in  the ",user_input," is ",sum)


    
```

```python
# reverse of the number (palandrome check) without using built-in functions.ex:-123!=321 ,121=121
number=int(input("enter the num that you want to reverse"))
original=number
rev=0
while(number>0):
    rem=number%10
    rev=rev*10+rem
    number=number//10
if original==rev:
    print("it is paldrome")
else:
    print("It is not palndrome")
```
